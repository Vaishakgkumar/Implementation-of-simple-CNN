 # **Implementation of Simple CNN**



This code creates a convolutional neural network (CNN) that can be used to classify images of dogs and cats. The CNN has four convolutional layers, each with 32, 64, 128, and 128 filters, respectively. The CNN also has three max pooling layers, each with a pool size of 2x2. The CNN is then flattened and connected to two dense layers, each with 128 and 64 neurons, respectively. The final layer is a 10-neuron softmax layer that predicts the probability of each class.

The CNN is compiled using the Adam optimizer and the sparse categorical crossentropy loss function. The model is trained for 10 epochs on the CIFAR-10 dataset. The model is then evaluated on the test set. The accuracy of the model on the test set is about 90%.

The model can be saved using the save() method. The model can then be loaded and used to classify new images.

# **Load The DATA**

![image.png](https://www.samyzaf.com/ML/cifar10/cifar1.jpg)

This code loads the CIFAR-10 dataset into memory. The CIFAR-10 dataset is a collection of 60,000 32x32 color images, divided into 10 classes, with 6,000 images per class. The code splits the dataset into a training set and a test set. The training set is used to train the model, and the test set is used to evaluate the model's performance.

The training images are 32x32 NumPy arrays, and the training labels are integers from 0 to 9, where each integer represents a class. The test images and test labels are the same format as the training images and labels.

The CIFAR-10 dataset is a popular dataset for machine learning research. It is a challenging dataset, but it is also a well-curated dataset. This makes it a good choice for beginners and experienced machine learning practitioners alike.

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
```

# **Normalize the data**


Code normalize the training and test data by dividing each pixel value by 255. This is important because it ensures that all of the data is in the same range, which helps the model to learn more effectively.

The reason why we need to normalize the data is because the pixel values in an image can vary greatly. For example, a pixel value of 255 represents the maximum brightness, while a pixel value of 0 represents the minimum brightness. If we do not normalize the data, the model will learn to pay more attention to pixels with high values, and less attention to pixels with low values. This is because the model will be able to more easily distinguish between pixels with high values, than between pixels with low values.

By normalizing the data, we ensure that all of the pixel values are in the same range. This helps the model to learn more effectively, because it can now focus on all of the pixels in the image, regardless of their value.

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

# **Creating the model**

![3](https://lh3.googleusercontent.com/yrHzday2CwSYLkXf9yKSoH-BpjqnnAuyiMvPAS5yS3-lFnl5jwkR6FoT_v2Vbi14s414fJSORuGLRQbHyYp6dtHDItRcSQnRWcd1JRGbZC5VlGTvH80gFZrHw8qg2Tx7ca2HYKFc)

Code creates a convolutional neural network (CNN) with 5 convolutional layers, 2 max pooling layers, 2 dense layers, and an output layer. The convolutional layers extract features from the input image, while the max pooling layers downsample the feature maps. The dense layers classify the features extracted by the convolutional layers.

The input layer has a shape of (32, 32, 3), which means that it expects an image with a width of 32 pixels, a height of 32 pixels, and 3 color channels (red, green, and blue). The convolutional layers have 32, 64, and 128 filters, respectively. The filters are used to extract features from the input image. The max pooling layers have a pool size of (2, 2), which means that they downsample the feature maps by a factor of 2. The dense layers have 128 and 64 neurons, respectively. The neurons in the dense layers classify the features extracted by the convolutional layers. The output layer has 10 neurons, which means that it can classify images into 10 different classes.

The activation function relu is used in the convolutional layers and the dense layers. The relu function is a non-linear function that helps the model to learn more complex features. The activation function softmax is used in the output layer. The softmax function normalizes the output of the output layer so that the sum of all the outputs is equal to 1. This helps the model to predict the probability of each class.

```python
model = tf.keras.Sequential([
  # Input layer
  tf.keras.layers.Input(shape=(32, 32, 3)),

  # Convolutional layer 1
  tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
  tf.keras.layers.MaxPooling2D((2, 2)),

  # Convolutional layer 2
  tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
  tf.keras.layers.MaxPooling2D((2, 2)),

  # Convolutional layer 3
  tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
  tf.keras.layers.MaxPooling2D((2, 2)),

  # Convolutional layer 4
  tf.keras.layers.Conv2D(256, (2, 2), activation='relu', padding='same'),
  tf.keras.layers.MaxPooling2D((2, 2)),

  # Flatten layer
  tf.keras.layers.Flatten(),

  # Dense layer 1
  tf.keras.layers.Dense(512, activation='relu'),

  # Dense layer 2
  tf.keras.layers.Dense(256, activation='relu'),

  # Output layer
  tf.keras.layers.Dense(10, activation='softmax')
])
```

# **Compile the model**

This code compiles the model using the Adam optimizer, the sparse categorical crossentropy loss function, and the accuracy metric. The Adam optimizer is a stochastic gradient descent optimizer that is known for its good performance. The sparse categorical crossentropy loss function is a loss function that is used for classification problems with a sparse output. The accuracy metric is a metric that measures the accuracy of the model.

The compile() method is used to configure the learning process of the model. The optimizer is used to update the weights of the model during training. The loss function is used to measure the error of the model. The metrics are used to evaluate the performance of the model.

The compile() method is a required method for all models. It must be called before the model can be trained.

The Adam optimizer is a good choice for most classification problems. It is a relatively simple optimizer to implement, and it is known for its good performance.

The sparse categorical crossentropy loss function is a good choice for classification problems with a sparse output. A sparse output is an output where the labels are integers. For example, if the model is classifying images of dogs and cats, the labels would be 0 for dogs and 1 for cats.

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

# **Train the model**

This code trains the model using the training data x_train and the training labels y_train for 10 epochs. An epoch is a complete pass through the training data. The model will learn from the training data and improve its accuracy over time.

The fit() method is used to train a model. The method takes three arguments: the training data, the training labels, and the number of epochs. The training data is a NumPy array or TensorFlow tensor that contains the training examples. The training labels are a NumPy array or TensorFlow tensor that contains the labels for the training examples. The number of epochs is an integer that specifies the number of times the model will iterate over the training data.

The fit() method is a required method for all models. It must be called to train the model.

```python
model.fit(x_train, y_train, epochs=20)
```

# **Evaluate the model**

This code evaluates the model using the test data x_test and the test labels y_test. The evaluation will produce a loss value and an accuracy value. The loss value measures the error of the model on the test data. The accuracy value measures the fraction of predictions that are correct.

The evaluate() method is used to evaluate a model. The method takes two arguments: the test data and the test labels. The test data is a NumPy array or TensorFlow tensor that contains the test examples. The test labels are a NumPy array or TensorFlow tensor that contains the labels for the test examples.

The evaluate() method is a required method for all models.

```python
model.evaluate(x_test, y_test)
```

# **Confusion Matrix**

The confusion matrix will show how many times the model correctly classified each class, as well as how many times it incorrectly classified each class. The accuracy of the model is then calculated by dividing the number of correct classifications by the total number of classifications.

For example, if the model has 100 predictions and it correctly classifies 90 of them, then the accuracy will be 90%.

The confusion matrix can be used to evaluate the performance of a classification model. It can be used to identify which classes the model is having difficulty with, and it can be used to see how the model's performance changes over time.

Here is a short note on the confusion matrix:

True Positives (TP) are the number of instances that were correctly classified as positive.
True Negatives (TN) are the number of instances that were correctly classified as negative.
False Positives (FP) are the number of instances that were incorrectly classified as positive.
False Negatives (FN) are the number of instances that were incorrectly classified as negative.

1. The accuracy of the model is calculated as follows:
  
>   **accuracy = (TP + TN) / (TP + TN + FP + FN)**

2. The precision of the model is calculated as follows:
  
>   **precision = TP / (TP + FP)**

3. The recall of the model is calculated as follows:

>   **recall = TP / (TP + FN)**

4. The F1 score of the model is calculated as follows:

>   **F1 = 2 * (precision * recall) / (precision + recall)**

```python
import numpy as np
from sklearn.metrics import confusion_matrix

# Get the predictions and actual labels
predictions = model.predict(x_test)
actual_labels = np.argmax(y_test, axis=1) # Convert to one-hot encoded format and get the indices of the highest values

# Create the confusion matrix
confusion_matrix = confusion_matrix(actual_labels, predictions.argmax(axis=1))

# Print the confusion matrix
print(confusion_matrix)

# Evaluate the classification
accuracy = np.sum(np.diagonal(confusion_matrix)) / np.sum(confusion_matrix)
print("Accuracy:", accuracy)
```

# **Ploting Images Using Matplot**
The code loads a pre-trained model that can classify images of cats and dogs.
* The code then gets the predictions from the model for a set of test images.
* The code then gets the labels for the test images.
* The code then creates a list of cat and dog images, based on the labels.
* The code then plots the cat and dog images, along with their accuracy scores.
* The code is well-formatted and easy to follow. It is also clear and concise. Overall, the code is well-written.

One thing that could be improved is the way the code handles the case where there are no cat or dog images in the test set. In this case, the code will generate an error. It would be better to handle this case gracefully, by printing a message or something similar.

**It is a good example of how to use TensorFlow to classify images.**

```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Load the cat and dog model
model = tf.keras.models.load_model('/kaggle/working/dog_cat_model.h5')

# Get the predictions from the model
predictions = model.predict(x_test)

# Get the labels from the test set
labels = y_test

# Create a list of cat and dog images
cat_images = []
dog_images = []
for i in range(len(predictions)):
  if labels[i] == 0:
    cat_images.append(x_test[i])
  else:
    dog_images.append(x_test[i])

# Plot the cat and dog images
plt.figure(figsize=(10, 10))
for i in range(4):
  plt.subplot(2, 2, i + 1)
  if i < len(cat_images):
    plt.imshow(cat_images[i].reshape(32, 32, 3), cmap='gray')
    plt.axis('off')
    plt.title(f'Image {i} | Accuracy: {predictions[i]}')
  else:
    plt.imshow(dog_images[i - len(cat_images)].reshape(32, 32, 3), cmap='gray')
    plt.axis('off')
    plt.title(f'Image {i} | Accuracy: {predictions[i]}')

plt.tight_layout()

plt.show()
```
